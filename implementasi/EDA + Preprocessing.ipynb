<h3> Drop Missing Value</h3>

```python
dataset_1 = dataset_1[dataset_1['title'].notna()]
dataset_1 = dataset_1[dataset_1['description'].notna()]
dataset_1 = dataset_1[dataset_1['content'].notna()]
```

```python
print("Cek Missing Value:\n" ,dataset_1.isnull().any())
dataset_1.isnull().sum()
```

```python
print("Dataset size:", dataset_1.shape)
print("Dataset columns:", dataset_1.columns)
```

```python
dataset_1['length'] =dataset_1['title'].apply(len)
```

```python
# view first 5 messages with length
dataset_1.head()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
Tweet = str(dataset_1.title.values)

wordcloud = WordCloud(
    width = 3000,
    height = 2000,
    background_color = 'black'
).generate(Tweet)

fig = plt.figure(
    figsize = (40, 30),
    facecolor = 'k',
    edgecolor = 'k')

plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

```python
# The length of the headline_text
lens = dataset_1.title.str.len()
print("Rata-rata panjang title:\n", lens.mean())
print("Standar Deviasi title:\n", lens.std())
print("Panjang minimal title:\n", lens.min())
print("Panjang maksimal title:\n", lens.max())

lens.hist();
```
